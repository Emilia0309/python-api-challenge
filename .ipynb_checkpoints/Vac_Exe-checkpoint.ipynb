```python
# Dependencies and Setup
import hvplot.pandas
import pandas as pd
import requests

# Import API key
from api_keys import geoapify_key
```

```python
# Read saved data
city_data_df = pd.read_csv("output_data/cities.csv", index_col="City_ID")

# Display sample data
city_data_df.head()
```

```python

import folium

# Create a base map
m = folium.Map(location=[0, 0], zoom_start=2)  # Centered at (0, 0)

# Define a function to choose color based on humidity level
def get_color(humidity):
    if humidity < 30:
        return 'green'
    elif 30 <= humidity < 60:
        return 'orange'
    else:
        return 'red'

# Loop through the DataFrame and add a marker for each city
for _, row in city_data_df.iterrows():
    folium.CircleMarker(
        location=(row['Lat'], row['Lng']),
        radius=row['Humidity'] / 10,  # Scale the humidity for visibility
        color=get_color(row['Humidity']),
        fill=True,
        fill_color=get_color(row['Humidity']),
        fill_opacity=0.6,
        popup=f"{row['City']} ({row['Country']}) - Humidity: {row['Humidity']}%"
    ).add_to(m)

# Create a legend
legend_html = '''
    <div style="position: fixed; 
                bottom: 50px; left: 50px; 
                width: 150px; height: 150px; 
                z-index: 9999; 
                font-size: 14px; 
                background-color: white; 
                border: 2px solid grey; 
                padding: 10px;">
    <b>Humidity Legend</b><br>
    <i style="color: green;"></i> Low (0-30%)<br>
    <i style="color: orange;"></i> Moderate (30-60%)<br>
    <i style="color: red;"></i> High (60-100%)<br>
    </div>
'''

m.get_root().html.add_child(folium.Element(legend_html))

# Save map to an HTML file
m.save("city_humidity_map.html")

# If you want to display the map in a Jupyter notebook
m
```

**Narrow down the city_data_df DataFrame to find your ideal weather condition**


```python
# Narrow down cities based on ideal weather conditions
ideal_conditions = city_data_df[
    (city_data_df['Max Temp'] >= 15) & (city_data_df['Max Temp'] <= 25) &
    (city_data_df['Humidity'] < 70) &
    (city_data_df['Wind Speed'] < 5)
]

# Drop any rows with null values (though there are none in this sample)
ideal_conditions = ideal_conditions.dropna()

# Display sample data
ideal_conditions.head()
```

**Step 3: Create a new DataFrame called hotel_df.**

```python
# Create original DataFrame
#ideal_df = pd.DataFrame(data)

# Use the Pandas copy function to create hotel_df
hotel_df = ideal_conditions[['City', 'Country', 'Lat', 'Lng', 'Humidity']].copy()

# Add an empty column, "Hotel Name," to the DataFrame
hotel_df['Hotel Name'] = ''

# Display sample data
hotel_df.head()
```

**Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates**

```python


import requests

# Set parameters to search for a hotel
radius = 10000  # Search radius in meters
geoapify_api_key = "6ffe16fb844345c9b1ba030b336cf435"  # Replace with your Geoapify API key

# Print a message to follow up the hotel search
print("Starting hotel search")

# Iterate through the hotel_df DataFrame
for index, row in hotel_df.iterrows():
    # Get latitude and longitude from the DataFrame
    lat = row['Lat']
    lng = row['Lng']
    
    # Set parameters for the API request
    params = {
        'apiKey': geoapify_api_key,
        'categories': 'accommodation.hotel',
        'filter': f'circle:{lng},{lat},{radius}',
        'bias': f'proximity:{lng},{lat}'
    }

    # Set base URL
    base_url = "https://api.geoapify.com/v2/places"

    # Make an API request using the params dictionary
    response = requests.get(base_url, params=params)

    # Convert the API response to JSON format
    try:
        name_address = response.json()
    except ValueError:
        print(f"Error decoding JSON response for {row['City']}")
        continue

    # Grab the first hotel from the results and store the name in the hotel_df DataFrame
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["features"][0]["properties"]["name"]
    except (KeyError, IndexError):
        # If no hotel is found, set the hotel name as "No hotel found".
        hotel_df.loc[index, "Hotel Name"] = "No hotel found"

    # Log the search results
    print(f"{hotel_df.loc[index, 'City']} - nearest hotel: {hotel_df.loc[index, 'Hotel Name']}")

# Display sample data
hotel_df



```

**Step 5: Add the hotel name and the country as additional information in the hover message for each city in the map**

```python
import folium



#hotel_df = pd.DataFrame(data)

# Create a base map
m = folium.Map(location=[0, 0], zoom_start=2)

# Add points to the map
for index, row in hotel_df.iterrows():
    # Create a popup message
    popup_message = f"{row['City']}, {row['Country']}<br>Nearest Hotel: {row['Hotel Name']}"
    
    # Add a marker with a popup to the map
    folium.Marker(
        location=[row['Lat'], row['Lng']],
        popup=popup_message,
        icon=folium.Icon(color='blue')  # You can customize the marker color
    ).add_to(m)

# Display the map
#m.save("hotel_map.html")  # Save the map to an HTML file
m  # If running in a Jupyter notebook, this will display the map inline
```
